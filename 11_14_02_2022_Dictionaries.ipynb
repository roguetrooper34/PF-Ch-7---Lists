```python
# Question 10
from random import shuffle
simulations = 2500000 # Every 500,000 simulations take approximately 12 seconds

flushes = 0

deck = [{'value': i, 'suit': c} for c in ["hearts", "diamonds", "spades", "clubs"] for i in range(2, 15)]
for i in range(simulations):
    shuffle(deck)
    Player1 = [deck[i] for i in range(5)]
    Player1 = [list(Player1[i].items()) for i in range(len(Player1))]

    flush = True
    for i in range (4):
        if Player1[i][1][1] != Player1[i+1][1][1]:
            flush = False
            break
    
    if flush == True: flushes += 1


print("Probablity of a flush in a five card hand: {0:.4%}".format(flushes/simulations))
```

```python
# Question 15
timezones = {"eastern": 0, "central": 1, "pacific": 3, "mountain": 2}
shorter_timezones = {"et": "eastern", "est": "eastern", "east" : "eastern", "eas": "eastern",
 "ct": "central", "cst": "central", "cen": "central", "cent": "central", 
 "mt": "mountain", "mst": "mountain", "mount": "mountain",
 "pt": "pacific", "pst": "pacific", "pac": "pacific", "pf": "pacific", "paf": "pacific"}

def verifyFormat(start_time):
    if (6 <= len(start_time) <= 8) and ('am' in start_time[-2:] or 'pm' in start_time[-2:]) and (":" in start_time[1] or ":" in start_time[2]):
        if start_time[0].isnumeric() and (1 <= int(start_time[0]) <= 12) and start_time[2:4].isnumeric() and (0 <= int(start_time[2:4]) <= 59):
            return 10
        elif start_time[0:2].isnumeric() and (1 <= int(start_time[0:2]) <= 12) and start_time[3:5].isnumeric() and (0 <= int(start_time[3:5]) <= 59):
            return 20
        else:
            return 0
    else:
        return 0 # 0 is for invalid format, 10 is for valid format with only 1 digit for hour, 20 is for valid format with two digits for the hour spot

def convertTime(start_time, start_zone, end_zone, code):
    if code == 10:
        hour = int(start_time[0])
        hourDifference = timezones[end_zone] - timezones[start_zone]
        if hour == 12 and (hour+hourDifference) < 12:
            hour += hourDifference
            if 'am' in start_time:
                start_time = str(hour) + start_time[1:-2] + 'pm'
            else:
                start_time = str(hour) + start_time[1:-2] + 'am'
        elif hour+hourDifference == 0:
            hour = 12
            start_time = str(hour) + start_time[1:]
        elif hour+hourDifference < 0:
            hour += hourDifference + 12
            if 'am' in start_time:
                start_time = str(hour) + start_time[1:-2] + 'pm'
            else:
                start_time = str(hour) + start_time[1:-2] + 'am'
        elif hour+hourDifference > 12:
            hour += hourDifference - 12
            if 'am' in start_time:
                start_time = str(hour) + start_time[1:-2] + 'pm'
            else:
                start_time = str(hour) + start_time[1:-2] + 'am'
        else:
            hour += hourDifference
            start_time = str(hour) + start_time[1:]
    else:
        hour = int(start_time[0:2])
        hourDifference = timezones[end_zone] - timezones[start_zone]
        if hour == 12 and (hour+hourDifference) < 12:
            hour += hourDifference
            if 'am' in start_time:
                start_time = str(hour) + start_time[2:-2] + 'pm'
            else:
                start_time = str(hour) + start_time[2:-2] + 'am'
        elif hour+hourDifference == 0:
            hour = 12
            start_time = str(hour) + start_time[2:]
        elif hour+hourDifference < 0:
            hour += hourDifference + 12
            if 'am' in start_time:
                start_time = str(hour) + start_time[2:-2] + 'pm'
            else:
                start_time = str(hour) + start_time[2:-2] + 'am'
        elif hour+hourDifference > 12:
            hour += hourDifference - 12
            if 'am' in start_time:
                start_time = str(hour) + start_time[2:-2] + 'pm'
            else:
                start_time = str(hour) + start_time[2:-2] + 'am'
        else:
            hour += hourDifference
            start_time = str(hour) + start_time[2:]
    return start_time

start_time = input("Please enter the current time (in the format 3:48pm): ")
verifyStatus = verifyFormat(start_time)
while verifyStatus == 0:
    print("Invalid format!")
    start_time = input("Please enter the current time (in the format 3:48pm): ")
    verifyStatus = verifyFormat(start_time)


start_zone = input("Please enter the initial time zone: ")
while start_zone not in timezones and start_zone not in shorter_timezones:
    print("Invalid starting time zone!")
    start_zone = input("Please enter the initial time zone: ")

end_zone = input("Please enter the ending time zone: ")
while end_zone not in timezones and end_zone not in shorter_timezones:
    print("Invalid desired time zone!")
    end_zone = input("Please enter the ending time zone: ")

if start_zone in shorter_timezones:
    start_zone = shorter_timezones[start_zone]
if end_zone in shorter_timezones:
    end_zone = shorter_timezones[end_zone]

end_time = convertTime(start_time, start_zone, end_zone, verifyStatus)
print("Time: {0}\nStarting zone: {1}\nEnding zone: {2}\n{3}".format(start_time, start_zone.title(), end_zone.title(), end_time))
```

```python
# Question 16
rom_dict = {"M": 1000, "CM": 900, "D": 500, "CD": 400, "C": 100, "XC": 90, "L": 50, "XL": 40, "X": 10, "IX": 9, "V": 5, "IV": 4, "I": 1}

def numeralToRoman(number):
    rom_number = ''
    while number > 0:
        for x, y in rom_dict.items():
            if number >= y:
                rom_number += x
                number -= y
                break
    return rom_number


def isRoman(number):
    number = set(list(number.upper()))
    for x in number:
        if x not in rom_dict:
            return False
    return True

def romanToNumeral(number):
    number = number.upper()
    orig_number = number
    biggest = 1000
    total_value = 0

    while len(number) > 0:
        if len(number) > 1:
            double_char = number[0:2]
        if double_char in rom_dict:
            value_double_char = rom_dict[double_char]
            if value_double_char > biggest:
                return 0
            else:
                total_value += value_double_char
                biggest = value_double_char
                number = number[2:]
        else:
            value_char = rom_dict[number[0]]
            if value_char > biggest:
                return 0
            else:
                total_value += value_char
                biggest = value_char
                number = number[1:]

    if numeralToRoman(total_value) == orig_number:
        return total_value
    else:
        return 0


def main():
    number = input("Please enter any number to convert: ")
    if number.isnumeric():
        number = int(number)
        while number < 1:
            print("Please enter a number above 1.")
            number = input("Please enter an integer to convert: ")
        print(numeralToRoman(number))

    else:
        if isRoman(number): valid_roman_value = romanToNumeral(number)
        while not isRoman(number) or valid_roman_value == 0:
            print("Invalid roman numeral!")
            number = input("Please enter a roman numeral: ")
            if isRoman(number): valid_roman_value = romanToNumeral(number)
        print(valid_roman_value)

main()
```
